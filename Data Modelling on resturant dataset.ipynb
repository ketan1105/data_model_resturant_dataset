# function to create database

```python
def create_database():
    conn = psycopg2.connect("host=localhost port=12345 dbname=postgres user=postgres password=1105")
    cur = conn.cursor()
    conn.set_session(autocommit=True)

    cur.execute("drop database if exists resturant")
    cur.execute("create database resturant")

    cur.close()
    conn.close()

    conn = psycopg2.connect("host=localhost port=12345 dbname=resturant user=postgres password=1105")
    cur = conn.cursor()
    return conn,cur
```

# Made a connection to the resturant database of postgres sql

```python
conn,cur = create_database()
```

# Created Profile table 

```python
profileData = """
create table if not exists profile(
userID varchar primary key,
latitude float,
longitude float,
smoker varchar,
drink_level varchar,
dress_preference varchar,
ambience varchar,
transport varchar, 
marital_status varchar,
hijos varchar,
birth_year integer,
interest varchar,
personality varchar,
religion varchar,
activity varchar,
color varchar,
weight integer,
budget varchar,
height float
);"""
```

```python
cur.execute(profileData)
conn.commit()
```

```python
insert_profileData = """
insert into profile
(userID, latitude, longitude, smoker, drink_level, dress_preference, ambience, transport, 
 marital_status, hijos, birth_year, interest, personality, religion, activity, color, 
 weight, budget, height)
values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);
"""

for i, row in user_profile_data.iterrows():
    cur.execute(insert_profileData, list(row))

conn.commit()
```

# Created payment table

```python
paymentData = """create table payment(
userID varchar references profile(userID),
Upayment varchar
);"""
```

# Some SQL query to show we have stored the data in Tables

# Query1:output the count of customers by each years

```python
q = """
select birth_year,count(*) from profile 
group by birth_year;
"""

cur.execute(q)
ans = cur.fetchall()
for row in ans:
    print(row)
```

# Query2:output the name of resturant visited by each user who give rating order by name of resturant in ascending order

```python
q = """
select r.userID,a.name from rating as r
join address as a on a.placeID = r.placeID
order by a.name;
"""

cur.execute(q)
ans = cur.fetchall()
for row in ans:
    print(row)
```
